# Notebook 01 — Activation Collection
Runs DINO ViT-B/16 on ImageNet val (first 10k images) and saves activations from
layers 4 and 12 (0-indexed: blocks[3] and blocks[11]) to `../data/`.

**Output:**
- `../data/layer4_activations.pt` — shape (1,970,000, 768)
- `../data/layer12_activations.pt` — shape (1,970,000, 768)

```python
import sys
sys.path.insert(0, '..')
from src.activation_collector import run_collection

IMAGENET_VAL_DIR = '/path/to/imagenet/val'  # ← change this
OUTPUT_DIR = '../data'
```

```python
run_collection(
    imagenet_val_dir=IMAGENET_VAL_DIR,
    output_dir=OUTPUT_DIR,
    layer_indices=[3, 11],
    n_images=10000,
    batch_size=64,
    device='cuda',
)
```

```python
import torch
acts4  = torch.load('../data/layer4_activations.pt', weights_only=False)
acts12 = torch.load('../data/layer12_activations.pt', weights_only=False)
print(f'Layer 4:  {acts4.shape}  mean={acts4.mean():.4f}  std={acts4.std():.4f}')
print(f'Layer 12: {acts12.shape}  mean={acts12.mean():.4f}  std={acts12.std():.4f}')
```
